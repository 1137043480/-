### 3.1 电影发展趋势

#### 电影数量发展

```python
# 1. 对时间字段进行处理
# apply
# lambda x: x[:4] 
# def func(x) 
#     return x:x[:4]
# def func(x):
#     return x*x 
# func = lambda x: x*x  
data['year'] = data['title_year'].apply(lambda x: x[:4]) 
data['month'] = data['title_year'].apply(lambda x: x[5:])  
data 
# 2. 根据年份对数据分组汇总 
movie_years_count = data.groupby('year')['movie_title'].count() 
movie_years_count
plt.figure(figsize=(20,8),dpi=80) 
movie_years_count.plot() 
plt.show()
```

#### 拓展 一图双Y轴
#### 月份对电影数量及票房影响

```python
# 1.引入字体
plt.rcParams['font.sans-serif'] = 'Arial Unicode MS' 

# 2.设置画布 
fig , ax1 = plt.subplots(figsize=(20,12))  

# 3.电影数量数据  
# 3.1电影数量数据及坐标
movie_month_count = data.groupby('month')['movie_title'].count() 
# movie_month_count  
# print(type(movie_month_count))
x = movie_month_count.index.astype('int')  
y1 = movie_month_count.values.astype('int')
# 3.2X轴
x_ticks_label =  ['{}月份'.format(i) for i in x]  
ax1.set_xticks(x) 
ax1.set_xticklabels(x_ticks_label) 
# 3.3y1轴 
y1_ticks= np.arange(0,max(y1),10) 
y1_ticks_label =  ['{}部'.format(i) for i in y1_ticks] 
ax1.set_yticks(y1_ticks) 
ax1.set_yticklabels(y1_ticks_label)  
# 3.4绘制电影数量 
ax1.plot(x,y1,'--',label='电影数量',color='g')
ax1.set_xlabel('月份') 
ax1.set_ylabel('数量')
ax1.plot(x,y1) 

# 4.电影票房数据  
# 4.1电影票房数据及坐标
movie_month_gross = data.groupby('month')['gross'].sum()/10000 
# 4.2X轴共用
ax2 = ax1.twinx() 
# 4.3y2轴  
y2 = movie_month_gross.values.astype('int')
# 4.4绘制电影票房
ax2.plot(x,y2,'-',label='电影票房', color='r')
# 4.5设置y2轴的刻度
y2_ticks = np.arange(0,max(y2),100000)  
y2_ticks_label =  ['{}万元'.format(i) for i in y2_ticks]  
ax2.set_yticks(y2_ticks) 
ax2.set_yticklabels(y2_ticks_label)  
ax2.set_ylabel('票房')

# 5.设置图例
handles1, labels1 = ax1.get_legend_handles_labels() 
handles2, labels2 = ax2.get_legend_handles_labels() 
plt.legend(handles1+handles2, labels1+labels2) 

# 6.显示 
plt.show()
```

### 3.2 电影情况分析 

#### 时长分析

```python
# 1.电影时长分布 
movie_duration = data.duration 
movie_duration 

# 2.创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 3.绘制直方图
# 3.1设置组距
distance = 5
# 3.2计算组数
group_num = int((max(movie_duration ) - min(movie_duration )) / distance)
# 3.3绘制直方图
plt.hist(movie_duration , bins=group_num)
# 3.4修改x轴刻度显示
plt.xticks(range(int(min(movie_duration)),int(max(movie_duration)))[::5])
# 3.5添加网格显示
plt.grid(linestyle="--", alpha=0.5)

# 3.6添加x, y轴描述信息
plt.xlabel("电影时长大小")
plt.ylabel("电影的数据量")

# 4.显示图像
plt.show()
```

#### 电影类型分析

```python
# 1.构建新的DataFrame,存储电影类型及票房
genre_data = pd.DataFrame(columns=['genre','gross']) 
genre_data  
# 2.读取data数据,拆分,按照行读取整个电影数 
# iterrows() 迭代器 行读取数据 索引,数据
for i,row_data  in data.iterrows(): 

    # 分割体裁类型 
    # Action|Adventure|Fantasy|Sci-Fi str类型
    # split list类型
    genres = row_data['genres'].split("|") 
#     print(genres)
    # 每一部电影体裁个数 
    n_genres = len(genres)    
#     print(n_genres)
    # 构建空字典,保存'genre'与'gross'
    dict_obj = {} 
    dict_obj['gross'] = [row_data['gross']] * n_genres 
    dict_obj['genre'] = genres   
    # 每一部电影的数据字典类型转化成DataFrame形式 
    genre_df = pd.DataFrame(dict_obj)  
    # 存储所有的电影电影类型及票房
    genre_data = genre_data.append(genre_df) 
# 3.数据写入到csv文件中   
genre_data.to_csv('genre_data.csv',index=None)
genre_data
```

```python
#按题材分类，统计电影数量 
genre_count = genre_data.groupby('genre').size().sort_values()
genre_count
plt.figure(figsize=(15.0, 10.0))
genre_count.plot(kind='barh')
plt.savefig('genre_count.png')
plt.show()
```

```python
#按题材分类，统计电影票房 
genre_gross = genre_data.groupby('genre')['gross'].sum()/10000/10000
plt.figure(figsize=(15.0, 10.0))
genre_gross.plot(kind='barh')
plt.savefig('genre_gross.png')
plt.show()
```
